```python
import pandas as pd

df = pd.read_csv("prompts.csv")
df
```

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(file_path="prompts.csv", source_column='prompt',)
prompts = loader.load()
prompts
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name='hkunlp/instructor-large')
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(prompts, embeddings)
vectorstore
```

```python
query = "suggest me the book name which is best for machine learing?"
docs = vectorstore.similarity_search(query)
docs
```

```python
retriever = vectorstore.as_retriever(
    search_kwargs={'k': 1}
)
print(retriever.get_relevant_documents(query)[0].page_content)
```
